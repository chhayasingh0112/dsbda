```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
l=sns.load_dataset("titanic")
l
```

```python
print(l.isnull().sum())
```

```python
l.info()
```

```python
l=l.drop(['deck'],axis=1)
l
```

```python
meanValue=l['age'].mean()
l['age'].fillna(value=meanValue, inplace=True)
```

```python
l.info()
```

```python
modeValue=l['embark_town'].mode()
l['embark_town'].fillna(value=modeValue, inplace=True)
```

```python
l.info()
```

```python
l["sex"].replace({'male':0,"female":1},inplace=True)
l
```

```python
just_dummies = pd.get_dummies(l['embarked'])

step_1 = pd.concat([l, just_dummies], axis=1)
step_1=step_1.drop(['S'],axis=1)
step_1
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(l['age'])
 
# Show plot
plt.show()
```
