```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = 'https://www.flipkart.com/thousand-splendid-suns/p/itmfbpxwhjphkgnr?pid=9781408844441&lid=LSTBOK9781408844441VFKV00&marketplace=FLIPKART&q=thousand+splendid+suns+book&store=bks&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&fm=personalisedRecommendation%2FC2&iid=eb02ccf9-6894-4817-b81b-505e26d06b46.9781408844441.SEARCH&ppt=pp&ppn=pp&ssid=74p3zwsuv40000001652861797407&qH=9e15981cc2440cce'
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('p',class_='_2sc7ZR _2V5EHH')
names
```

```python
cust_name = []
for i in range (0,len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

```python
title = soup.find_all('p',class_='_2-N8zT')
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
rating = soup.find_all('div',class_='_3LWZlK _1BLPMq')
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
review = soup.find_all("div",class_='t-ZTKy')
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['names']=cust_name
df['Review Title']=review_title
df['Ratings']=rate
df['Review']=review_content
```

```python
df
```

```python
df.to_csv('flipkart_reviews.csv', index=False, encoding='utf-8')
```
